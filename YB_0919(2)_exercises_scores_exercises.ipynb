````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

df = pd.DataFrame(raw_data)
df
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(10,8))
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'], alpha=0.8, color='green')

plt.xlabel('Pre Test Score', fontsize=12)
plt.ylabel('Post Test Score', fontsize=12)
plt.title('Scatterplot of Pre vs Post Test Scores (Size: Age)', fontsize=14)

for i, row in df.iterrows():
    plt.text(row['preTestScore']+0.5, row['postTestScore']+0.5, row['first_name'], fontsize=9)

plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize=(10,8))
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['postTestScore']*4.5, c=df['female'], alpha=0.8)

plt.xlabel('Pre Test Score', fontsize=12)
plt.ylabel('Post Test Score', fontsize=12)
plt.title('Scatterplot of Pre vs Post Test Scores (Size ~ PostTestScore, Color ~ Sex)', fontsize=15)

cbar = plt.colorbar()
cbar.set_label('Female (0=Male, 1=Female)')

plt.show()
````

### BONUS: Create your own question and answer it.

preTestScore 대비 postTestScore에서 점수가 가장 많이 오른 순서대로 barplot 출력하기

````python
df['score_diff'] = df['postTestScore'] - df['preTestScore']
df.sort_values(by='score_diff', ascending=False, inplace=True)

plt.figure(figsize=(8,5))
plt.bar(df['first_name'], df['score_diff'], color="orange")
plt.ylabel("Score Improvement")
plt.xlabel("Student")
plt.title("Improvement from PreTest to PostTest")
plt.show()
````
