## 주제

라이프로그 데이터를 활용한 수면 품질 및 상태 예측

## 데이터

### 데이터 설명

- Q1: 취침 후 수면의 질
- Q2: 취침 전 피로도
- Q3: 취침 전 스트레스
- S1: 총 수면시간
- S2: 수면효율
- S3: 수면지연시간

## 코드 흐름

1. 시계열 기반 피처 생성 함수 생성
2. parquet 파일 전처리 및 피처 집계
3. 경로 설정
4. 피처 생성
5. 학습 데이터 로딩
6. 날짜 피처 생성
7. 테스트 데이터
8. 피처/라벨 분리
9. 결측치 처리
10. 학습 및 예측
11. 저장

### 모델링

- LightGBM + CatBoost 앙상블
- 각 target을 개별 모델로 학습
- 교차 검증
    - StratifiedKFold 5폴드
    - fold마다:
        - LightGBM 확률 예측
        - CatBoost 확률 예측
        - 두 모델의 확률은 3:7 가중 평균
- 예측 결과
    - 전체 폴드의 예측을 합산 후 평균
    - 최종적으로 argmax로 가장 높은 확률의 class 선택

## 차별점 및 배울점

- 다양한 시계열 기반 피처를 생성함
- subject_id 별로 ‘시계열→통계 집계’ 구조로 과적합을 낮추면서도 패턴을 담음
- LightGBM + CatBoost 앙상블에서 가중치를 다르게 줌
- Stratified K-Fold로 각 target을 따로 학습
